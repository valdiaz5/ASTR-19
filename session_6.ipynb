```python
import numpy as np
from astropy.table import Table
```

This function returns sin(x)

```python
def sin_data(x):
    sinx = np.sin(x)
    return sinx
```

This function returns cos(x)

```python
def cos_data(x):
    cosx= np.cos(x)
    return cosx
```

This functon tabulates sinx and cosx vs x between 0 and 2pi with 1000 enteries 

```python
x = np.linspace(0, 2*np.pi, 1000)
sin_values = sin_data(x)
cos_values = cos_data(x)

table = Table([x, sin_values, cos_values],names = ("x", "sin(x)", "cos(x)"))
#print (table)
```

This cell uses a for loop to print out the first 10 enteries of x, sinx, and cosx, in columns 

```python
for i in range(10):
    print(f'{x[i]}, {sin_values[i]}, {cos_values[i]}')
    
    
```
