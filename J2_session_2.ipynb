```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)
def func(x, a, b):
    y = a*np.exp(-b*x)
    return y
a = 2
b = 5
```

```python
y = func(x, a, b)

plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
sigma = np.linspace(2, 0.2, len(y))

scatter = np.random.normal(loc=0, scale=sigma)
    
y_scatter = y + scatter

plt.figure()
plt.errorbar(x, y_scatter, yerr=sigma, fmt = 'o', color = 'olivedrab', label = "Scatter")

plt.plot(x, y, color='red', label="$y=ae^{-bx}$")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.savefig("session_2.pdf", bbox_inches = 'tight')
plt.show()
```
