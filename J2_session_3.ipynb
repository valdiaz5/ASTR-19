```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import warnings
```

Prompt 1

```python
x = np.linspace(0, 2*np.pi, 100)
def func(x, a, b):
    y = a*np.exp(-b*x)
    return y
a = 2
b = 5
```

```python
y = func(x, a, b)

plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

Prompt 2

```python
sigma = np.linspace(2, 0.2, len(y))

scatter = np.random.normal(loc=0, scale=sigma)
    
y_scatter = y + scatter

plt.figure()


plt.errorbar(x, y_scatter, yerr=sigma, fmt = 'o', color = 'olivedrab', label = "Scatter")

plt.plot(x, y, color='red', label="$y=ae^{-bx}$")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.savefig("session_2.pdf", bbox_inches = 'tight')
plt.show()
```

Prompt 3

```python
def model_1(x, a, b):
    y = a*np.exp(-b*x)


    return y

def model_2(x, a, b):
    y = a/(x**b)
    return y 
```

```python
#find the best fit values 
popt_1, pcov_1 = curve_fit(model_1, x, y_scatter, p0=[2, 6])
popt_2, pcov_2 = curve_fit(model_2, x[1:], y_scatter[1:], p0 =[1, 2], )

#assign best fit values to the models 
best_fit_1 = model_1(x, popt_1[0], popt_1[1])
best_fit_2 = model_2(x[1:], popt_2[0], popt_2[1])
```

```python

#bring best fit values 
print(f"Model 1 function best fit values: a = {popt_1[0]:.4f}, b ={popt_1[1]:.4f}")
print(f"Model 2 function best fit values: a = {popt_2[0]:.4f}, b ={popt_2[1]:.4f}")

#plot scatter, line, and best fit lines
plt.figure()
plt.errorbar(x, y_scatter, yerr=sigma, fmt = 'o', color = 'pink', alpha = 0.7,label = "Scatter")

plt.plot(x, y, color='tomato', label="$y=2e^{-5x}$")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x, best_fit_1, color = 'lightseagreen', label = "Model 1 Best Fit $y=ae^{-bx}$")
plt.plot(x[1:], best_fit_2, color = 'darkorchid', label = "Model 2 Best Fit $y=a/x^b$")

plt.legend()
plt.savefig("session_3.pdf", bbox_inches = 'tight')
plt.show()
```
