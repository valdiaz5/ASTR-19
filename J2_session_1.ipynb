```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
def func(x):
    y = a*np.exp(-b*x)
    return y
a = 2
b = 5
```

```python
plt.figure()
plt.plot(x, func(x))
plt.xlabel("x")
plt.ylabel("y")

plt.savefig("session_1.pdf", bbox_inches = 'tight')
```
