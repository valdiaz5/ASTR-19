```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3
values = (np.random.pareto(a, 1000) + 1)
```

```python
t = plt.hist(values, bins = 100, density=True)
plt.xlabel("Means of Pareto Distibution")
plt.ylabel("Probability Density")
plt.savefig("session_4.pdf", bbox_inches = 'tight')
```
